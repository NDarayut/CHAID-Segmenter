```python
import pandas as pd
from chaid_segmenter import ChaidSegmenter
```

```python
data = pd.read_csv("sample_loan_book.csv")
```

```python
data.head()
```

```python
seg = ChaidSegmenter(
    target="90PLSDPD_MOB6",
    positive_class=1,
    predictors={
        "PRODUCT_TYPE": "nominal",
        "TENURE_QUANTITY": "target",
        "DISBURSEMENT_AMOUNT": "target",
        "MEMBER_ID": {"method": "target", "categorical": True},   # group as an id, not a number
    },
    max_depth=3,
    min_child_node_size=0.02,
).fit(data)
```

```python
seg.summary().to_csv("summary.csv", index=False)
```

```python
seg.plot("tree.png")
```

```python
import matplotlib.pyplot as plt
fig = seg.plot()
plt.show()
```
